Classification Tree

 - sequence of if-else question about individual features
 - Objective: infer class label
 - Able to capture non-linier relationship between features and label
 - Dont required feature scaling(ex: standarization)

```python
import pandas as pd
import numpy as np


breast_cancer_data = pd.read_csv('DesicionTree/data.csv')
# breast_cancer_data.dropna(axis=1, thresh=None) 
breast_cancer_data = breast_cancer_data.dropna(axis='index', how='all') 
# breast_cancer_data['Unnamed'].isnull().sum()

# breast_cancer_data['diagnosis'] = pd.to_numeric(breast_cancer_data.diagnosis, errors='coerce')

breast_cancer_data.rename( columns={'Unnamed: 32':'Unknown_Column'}, inplace=True )
breast_cancer_data['Unknown_Column'] = breast_cancer_data['Unknown_Column'].astype(float)
breast_cancer_data = breast_cancer_data.dropna(axis=1)

breast_cancer_data['diagnosis'].replace(to_replace=['M','B'], value=[0,1],inplace=True)
breast_cancer_data.tail()
```

```python
from sklearn.model_selection import train_test_split

# Create arrays for the features and the target: X, y
X, y = breast_cancer_data.iloc[:,:-1], breast_cancer_data.iloc[:,-1]
# X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, stratify=y, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33,  random_state=42)
breast_cancer_data.dtypes
```

```python
# Import DecisionTreeClassifier from sklearn.tree
from sklearn.tree import DecisionTreeClassifier
# Instantiate a DecisionTreeClassifier 'dt' with a maximum depth of 6
dt = DecisionTreeClassifier(max_depth=6, random_state=1)

# Fit dt to the training set
dt.fit(X_train, y_train)

# Predict test set labels
y_pred = dt.predict(X_test)
print(y_pred[0:5])
```

```python
# Import accuracy_score
from sklearn.metrics import accuracy_score

dt = DecisionTreeClassifier(max_depth=6, random_state=1)
# Fit dt to the training set
dt.fit(X_train, y_train)

# Predict test set labels
y_pred = dt.predict(X_test)



# Compute test set accuracy  
acc = accuracy_score(y_pred, y_test)
print("Test set accuracy: {:.2f}".format(acc))
```
