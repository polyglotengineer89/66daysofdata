# Histogram

```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

df_canada = pd.read_csv("data_visual_1.csv")
df_canada['2013'].plot(kind='hist')

plt.title("Histogram of Imigration from several country")
plt.ylabel("Number of country")
plt.xlabel("Number of imigration")

plt.show()
```

```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np

df_canada = pd.read_csv("data_visual_1.csv")

count, bin_edges = np.histogram(df_canada['2013'])

df_canada['2013'].plot(kind='hist', xticks=bin_edges)

plt.title("Histogram of Imigration from several country")
plt.ylabel("Number of country")
plt.xlabel("Number of imigration")

plt.show()
```

# Bar Chart

```python
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

import numpy as np

df_canada = pd.read_csv("data_visual_1.csv")

years = list(map(str, range(1980, 2014)))

df_iceland = df_canada.loc[df_canada['OddName']=='Algeria', years]

df_iceland.plot(kind="bar")

plt.title('Algeria immigrate to canada from 1980 to 2013')
plt.xlabel('years')
plt.ylabel('Number of immigrant')

plt.show()
```

```python
df_canada['2013'].plot(kind='bar', rot=90)
```

```python
import matplotlib.pyplot as plt

transparency = 0.35 
df_canada.plot(kind='area', alpha=transparency, figsize=(20, 10))

plt.title('Plot Title')
plt.ylabel('Vertical Axis Label')
plt.xlabel('Horizontal Axis Label')

plt.show()
```

# Pie Chart

```python
df_canada['Total'].plot(kind='pie')

plt.title('Imigration to canada by continent')
plt.show()
```

```python
df_afgan = pd.read_csv("data_visual_afgan.csv")

df_afgan_new = df_afgan['Country'].loc[['Afganistan'], years].transpose()

df_afgan_new.plot(kind='box')

plt.title('Imigration Afganistan to canada by continent')
plt.ylabel("Number of Immigrant")
plt.show()
```
